## Visualizations


```python
penguins = sns.load_dataset("penguins")
penguins
```

### Draw a barplot by specie

 - What represents the vertical axis?
 - Write and run a code that outputs the plot below

```python
# Group the data by 'species' and calculate the mean for each group
grouped = penguins.groupby('species')['body_mass_g'].mean()

# Plot a barplot
plt.bar(grouped.index, grouped.values)
plt.xlabel('Specie')
plt.ylabel('Mean Value')
plt.title('Barplot by Specie')
plt.show()
```

The vertical axis represents the Mean Value

### Draw a histogram

 - Which is the difference between histogram and barplot?
 - Write and run a code that outputs the plot below

```python
species_counts = penguins['species'].value_counts()
species_names = species_counts.index

plt.bar(species_names, species_counts)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Histogram of Penguin Species')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

A histogram is used to visualize the distribution of a single continuous variable by dividing its range into bins and displaying the frequency or density of data points in each bin, while a barplot is employed to compare different categories or groups, representing their values with rectangular bars, typically showing counts or means, making it suitable for comparing data across categories or groups.

### Scatter plots

 - What does represent each data point?
 - Write and run a code that outputs the plots below

```python
import pandas as pd
import matplotlib.pyplot as plt

# We are going to create scatter plots for 'bill_length_mm' vs. 'bill_depth_mm'

plt.figure(figsize=(8, 4))

# Scatter plot with red dots
plt.scatter(penguins['bill_length_mm'],penguins['bill_depth_mm'], color='red', label='Data Points')

plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.title('Scatter Plot: Bill Length vs. Bill Depth')
plt.grid(True)
plt.legend()
plt.show()
```

Each data point in a scatter plot represents a single observation or data record in a dataset. In a scatter plot, the horizontal (x-axis) and vertical (y-axis) positions of each data point are determined by the values of two different variables. This allows us to visualize the relationship or association between these two variables. Scatter plots are particularly useful for identifying patterns, trends, correlations, or clusters in data.

### Lineplot

 - Lineplots are suitable for which type of data?
 - Write and run a code that outputs the plot below

Lineplots are suitable for visualizing data that represents a trend or relationship between two continuous variables, typically with data points that are connected by line segments. They are particularly useful for showing how a variable changes over time or another continuous scale.

```python
stocks = pd.read_csv('stocks.csv', sep = ';')
stocksMMM = stocks[(stocks['ccode'] == 'MMM') & (stocks['Date'] <= '2021-01-30')]
stocksMMM
```

```python
# Convert the 'Date' column to a datetime type
stocksMMM['Date'] = pd.to_datetime(stocksMMM['Date'])

plt.figure(figsize=(10, 6))

# Line plot for 'Close' prices
plt.plot(stocksMMM['Date'], stocksMMM['Close'], label='Close', color='blue', marker='o', linestyle='-')

plt.xlabel('Date')
plt.ylabel('Close')
plt.title('Line Plot: Close Over Time')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

### Boxplot


 - Can you describe the different parts of the boxplot?
 - Code the plot below
 - Which is the relationship between the boxplot and the table below?

Here are the different parts of a boxplot:

1.Box: The box represents the interquartile range (IQR), which spans from the 25th percentile (Q1) to the 75th percentile (Q3) of the data. The length of the box is a visual representation of the spread of the middle 50% of the data.

2.Median (Line Inside the Box): The line inside the box represents the median (Q2), which is the middle value of the dataset when it is ordered from lowest to highest.

3.Whiskers: The whiskers extend from the box to the minimum and maximum values within a certain range from the quartiles. By default, the whiskers typically extend to 1.5 times the IQR. Data points beyond this range are considered outliers and are plotted individually as points beyond the whiskers.

4.Outliers (Individual Points): Outliers are data points that fall outside the whiskers and are plotted individually as points. They are potential anomalies or extreme values in the dataset.

5.Notches (Optional): Notches on the sides of the boxplot can be used to display a confidence interval around the median. If the notches of two boxplots do not overlap, it suggests that there may be a significant difference in the medians of the two groups.

Relationship Between Boxplot and Table:

A boxplot is a visual representation of the key summary statistics of a dataset, allowing you to quickly assess the central tendency, spread, and presence of outliers. It provides a compact and informative summary of the data's distribution. In contrast, a table typically presents numerical values and statistics, such as mean, median, standard deviation, and quartiles, in a tabular format. The boxplot complements the table by providing a visual depiction of the same statistics, making it easier to understand the data's characteristics and identify potential outliers or skewness. Together, the boxplot and the table provide a comprehensive view of the dataset's distribution and summary statistics.

```python
# Create a boxplot for 'Close'
plt.figure(figsize=(8, 6))
plt.boxplot(stocksMMM['Close'], vert=False)  # vert=False for a horizontal boxplot
plt.xlabel('Close')
plt.title('Boxplot of Close')
plt.grid(True)
plt.show()
```
